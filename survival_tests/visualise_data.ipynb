```python
from fetch_dataframe_from_server import fetch_dataframe_from_server

from matplotlib import pyplot as plt
```

```python
metric_name = 'KendallTau'

sql_statement = """SELECT scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria, AVG(result) as result FROM test_more_approaches
WHERE scenario_name = 'QBF-2016' AND metric = 'KendallsTau_b' AND stopping_criteria = 'max_depth_3'
GROUP BY scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria """
    
df_decision_trees = fetch_dataframe_from_server(sql_statement)


sql_statement = (
        """SELECT scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria, AVG(result) as result FROM test_more_approaches
WHERE scenario_name = 'QBF-2016' AND metric = 'KendallsTau_b' and approach = 'sbs'
GROUP BY scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria """
    )
sbs_baseline = fetch_dataframe_from_server(sql_statement)

sql_statement = (
            """SELECT scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria, AVG(result) as result FROM test_more_approaches
WHERE scenario_name = 'QBF-2016' AND metric = 'KendallsTau_b' and approach = 'per_algorithm_DecisionTreeRegressor_regressor'
GROUP BY scenario_name, approach, metric, impact_factor, ranking_error, borda_score, stopping_criteria """
)
decision_tree_regressor_baseline = fetch_dataframe_from_server(sql_statement)
```

```python
fig, axs = plt.subplots()

axs.plot(df_decision_trees['impact_factor'], df_decision_trees['result'], 'o', label = 'lambda impact' )
axs.axhline(y = decision_tree_regressor_baseline['result'].iat[0], color = 'r', linestyle = '--', label = 'regression tree regressor baseline')
axs.axhline(y =sbs_baseline['result'].iat[0], color = 'g', linestyle = '--', label = 'sbs baseline')
fig.suptitle("Kendall's Tau", fontsize=16)
axs.set_xlabel('regression tree regressor baseline')
axs.set_ylabel('Kendalls Tau')
plt.legend(loc=(0.2,-0.4))
```
